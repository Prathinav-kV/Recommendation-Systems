### Collaborative Filtering Methods

```python
from surprise import Dataset,Reader

reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(super_df[['author', 'product', 'score']], reader)
data
```

```python
from surprise.model_selection import train_test_split
trainset, testset = train_test_split(data, test_size=.25,random_state=123)
```

```python
from surprise import SVD, KNNWithMeans, accuracy
```

```python
svd_model = SVD(n_factors=5,biased=False)
svd_model.fit(trainset)
```

```python
test_pred = svd_model.test(testset)
```

```python
# compute RMSE
accuracy.rmse(test_pred)
```

```python
from surprise.model_selection import cross_validate
cross_itembased = cross_validate(svd_model, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
# Item based KNNWithMeans
```

```python
knn_i = KNNWithMeans(k=10, sim_options={ 'user_based': False})
knn_i.fit(trainset)
```

```python
test_pred=knn_i.test(testset)
accuracy.rmse(test_pred)
```

```python
from surprise.model_selection import cross_validate
cross_itembased = cross_validate(knn_i, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
# User based KNNWithMeans
```

```python
knn_u = KNNWithMeans(k=10, sim_options={ 'user_based': True})
knn_u.fit(trainset)
```

```python
test_pred=knn_u.test(testset)
accuracy.rmse(test_pred)
```

```python
from surprise.model_selection import cross_validate
cross_itembased = cross_validate(knn_u, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

### Here we see that the SVD is the best collaborative filtering model for the recommendation
### as compared to KNNWithMeans item and user based

### Q10. In what business scenario you should use popularity based Recommendation Systems?

Popularity based recommender system is used when the user requires the most popular items, news, and information.
In a case where the user is looking for a Television, and the user would like suggestions - then the most popularly bought 
TV would be recommended based on the purchase history, the reviews and how many users have clicked on it. 

This form of recommendation is not personalized, and doesn't suffer from cold start.

### Q11. In what business scenario you should use CF based Recommendation Systems?

Collaborative Filtering based Recommender systems are used when the user requires personalized suggestions for the service or product they require. Let us consider that the user wants to watch a movie; here if the system understands the user's taste in movies, then it can recommend movies similar to his preference. CF based recommender systems do the very the same, by understanding what the user's preference is and suggesting products or services which they would like.

This system suffers from cold start issue whenever a new user or a new item is considered.
